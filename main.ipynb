```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.mobilenet import preprocess_input
from tkinter import filedialog
import tkinter as tk

# Load the trained model
model = load_model('/Users/chamidiperera/Documents/Final year/Final Year Project/Project CARE/CARE/MobileNetV2.h5')

# Function to preprocess each frame
def preprocess_frame(frame):
    resized_frame = cv2.resize(frame, (224, 224))
    preprocessed_frame = preprocess_input(resized_frame)
    return np.expand_dims(preprocessed_frame, axis=0)

# Function to display predictions on the frame
def display_predictions(frame, predictions, threshold=0.5):
    label = "NormalEyes" if predictions[0][0] < threshold else "EyeLidTumor"
    confidence = predictions[0][0] if label == "EyeLidTumor" else 1 - predictions[0][0]
    color = (0, 0, 255) if label == "EyeLidTumor" else (0, 255, 0)
    cv2.putText(frame, f'Prediction: {label}, Confidence: {confidence:.2f}', (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 1, cv2.LINE_AA)

# Function to select an image file using a file dialog
def select_image_file():
    root = tk.Tk()
    root.withdraw()
    file_path = filedialog.askopenfilename()
    return file_path

# Load image file
image_path = select_image_file()
if image_path:
    frame = cv2.imread(image_path)
    processed_frame = preprocess_frame(frame)
    predictions = model.predict(processed_frame)
    display_predictions(frame, predictions)
    cv2.imshow('Image Prediction', frame)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
else:
    print("No image selected.")
```
